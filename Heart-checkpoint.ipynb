```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=20,stratify=y)
print ("train feature shape: ", x_train.shape)
print ("test feature shape: ",x_test.shape)
```

```python
from sklearn.svm import SVC

from sklearn.model_selection import GridSearchCV

from sklearn import preprocessing 
from sklearn.preprocessing import StandardScaler
```

```python
pipe_steps = [('scaler', StandardScaler()), ('pca', PCA()), ('SupVM', SVC(kernel='rbf'))]

check_params= {
    'pca__n_components': [2], 
    'SupVM__C': [0.1, 0.5, 1, 10,30, 40,41,42,43,46,49, 50, 75, 100, 500, 1000], 
    'SupVM__gamma' : [0.001, 0.005, 0.01, 0.05, 0.07, 0.1, 0.5, 1, 5, 10, 50]
}
```

```python
from sklearn.pipeline import Pipeline

pipeline = Pipeline(pipe_steps)
```

```python
# for CV=4 only
from sklearn.model_selection import GridSearchCV
gridx=GridSearchCV(pipeline,param_grid=check_params,cv=5)
gridx.fit(x_train,y_train)


print("score = %3.2f" %(gridx.score(x_test,y_test)))
print (gridx.best_params_)
```

```python
# I love You So Much 
from tqdm import tqdm_notebook as tqdm
```

```python
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings("ignore")

print("Start Fitting Training Data")
for cvx in tqdm(range(2,6)):
    newgrid=GridSearchCV(pipeline,param_grid=check_params,cv=cvx)
    newgrid.fit(x_train,y_train)
    print("Score for %d fold CV = %3.2f"%(cvx,newgrid.score(x_test,y_test)))
    print ("!!!!!!!! Best-Fit Parameters From Training Data !!!!!!!!!!!!!!")
    print (newgrid.best_params_)
    print(" ")
    print(" ")
    
print("Out of Loop")
```
